```python
# menampilkan dataset asli

df = pd.read_csv('/content/room_price_data.txt')
df
```

```python
#memisahkan data untuk train dan test kemudian menampilkannya

df_train = df.head(36)    #data atas
df_test = df.tail(9)      #data bawah

X_train = df_train['tourist'].values
Y_train = df_train['price'].values

X_test = df_test['tourist'].values
Y_test = df_test['price'].values

df
```

```python
# Membuat plot data points untuk tourist (x) dengan price (y)

X = df['tourist'].tolist()
Y = df['price'].tolist()

plt.scatter(X, Y, label='Data Points', color='#663399',)
plt.xlabel("tourist")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
#Deklarasi variabel

n = len(X_train)
XY_sum = 0
Y2_sum = 0
Y_sum = 0
X2_sum = 0
X_sum = 0
```

```python
# Mencari nilai a sebagai Y-intersept dan b sebagai Slope

for i in range(n):
    XY_sum += X_train[i]*Y_train[i]
    Y_sum += Y_train[i]
    Y2_sum += Y_train[i]**2
    X_sum += X_train[i]
    X2_sum += X_train[i]**2
a = (Y_sum*X2_sum - X_sum*XY_sum)/(n*X2_sum - X_sum**2)
b = (n*XY_sum - X_sum*Y_sum)/(n*X2_sum - X_sum**2)


print(f"Y-intersept = {a}")
print(f"Slope = {b}")
```

```python
# Persamaan regresi 

X = np.linspace (np.min(X_train), np.max(X_train))
Y = a + b*X

# Membuat plot data train dan regresi linearnya.

plt.plot(X, Y, color='#FF0000',label='Linear Regression')
plt.scatter(X_train, Y_train, color='#663399', label='Data Points')
plt.xlabel("tourist")
plt.ylabel("price")
plt.legend()
plt.show()
```
